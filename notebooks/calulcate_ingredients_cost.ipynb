```python
sizes_dict = {'s': "Weight (in grams)", 'm': "M (S*1.5)", 'l': "L(S*2)", 'xl': "XL(S*2.5)", 'xxl': "XXL(S*3)"}

ingredients_cost = {}

pizza_weight = {}


for original_name, general_name in pizza_names_dict.items():

    ingredients_df = data_sheet2[data_sheet2['pizza_name'] == general_name]
    temp_size = original_name.split("_")[-1]
    temp_df = ingredients_df[["ingredients", sizes_dict[temp_size], "Price in SAR(max)"]]
  
    temp_df['cost_per_ingredient'] = ingredients_df[sizes_dict[temp_size]] / 555 * temp_df["Price in SAR(max)"]

    ingredient_wieght_temp = temp_df.set_index('ingredients')[sizes_dict[temp_size]].to_dict()
    cost_pre_ingredient_dict = temp_df.set_index('ingredients')['cost_per_ingredient'].to_dict()

    for key, val in ingredient_wieght_temp.items():
        if key in pizza_weight.keys():
            pizza_weight[key] += float(val)
            ingredients_cost[key] += float(cost_pre_ingredient_dict[key])

        else:
            pizza_weight[key] = float(val)
            ingredients_cost[key] = float(cost_pre_ingredient_dict[key])


    # ingredients_cost[original_name] =  float(temp_df['cost_per_ingredient'].sum())
    
```

```python
pizza_weight_df = pd.DataFrame(pd.Series(pizza_weight).to_frame().T)
pizza_weight_df = pizza_weight_df.T

# convert to KG
pizza_weight_df[0] = pizza_weight_df[0] / 1000 * 2
pizza_weight_df.columns = ["Weight (K/G)"]
pizza_weight_df['Ingredients'] = pizza_weight_df.index

pizza_weight_df = pizza_weight_df[['Ingredients', "Weight (K/G)"]]
pizza_weight_df = pizza_weight_df.reset_index(drop=True)

pizza_weight_df = pizza_weight_df.sort_values("Weight (K/G)", ascending=False)
pizza_weight_df
```

```python
data_sheet2.head(2)
```

```python
ingredients_cost_df = data_sheet2.groupby(['ingredients'], as_index=False).agg({'Price in SAR(max)': 'max'})
ingredients_cost_df = ingredients_cost_df.rename({'ingredients': 'Ingredients', "Price in SAR(max)": 'Price per KG'}, axis=1)
ingredients_cost_df
```

```python
merged_df = pd.merge(pizza_weight_df, ingredients_cost_df, on="Ingredients", how="inner")
merged_df['Price per KG'] = merged_df['Price per KG'].apply(lambda x: round(x, 2))
merged_df['Weight (K/G)'] = merged_df['Weight (K/G)'].apply(lambda x: round(x, 2))
merged_df = merged_df.rename({"Price per KG": "Price per KG (SAR)"}, axis=1)
merged_df
```

```python
merged_df.to_csv("../data/processed/Ingredients_Weights.csv", index=False)
```
